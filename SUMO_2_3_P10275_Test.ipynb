# Load the Trained SUMO2/3 Model which is trained on Hendriks et al. SUMO2/3 Dataset

```python
model_SUMO_2_3 = keras.models.load_model("SUMO2_330482327.h5")
```

# Load the Test Protein

```python
df_tasnem_next_protein = pd.read_csv("P10275_Prot_Trans_.csv",header=None)

Header_Tasnem = ["Lysine"]
features = [str(i) for i in range(1,1025)]
Header = Header_Tasnem+features

df_tasnem_next_protein.columns = Header
```

```python
df_only_K = df_tasnem_next_protein.loc[df_tasnem_next_protein["Lysine"] == "K"]
```

```python
TASNEM_NEXT_PROTEIN = np.array(df_only_K.iloc[:,1:])
```

```python
Y_pred_TASNEM = model_SUMO_2_3.predict(TASNEM_NEXT_PROTEIN)
Y_pred_TASNEM_Ones = Y_pred_TASNEM[:,1]
Y_pred_TASNEM_Ones
```

```python
Y_pred_TASNEM_Ones_T_F = Y_pred_TASNEM_Ones>0.5
```

```python
df_only_K["Confidence"] = Y_pred_TASNEM_Ones
```

```python
df_only_K["Status"] = Y_pred_TASNEM_Ones_T_F 
```
