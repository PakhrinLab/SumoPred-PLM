```python
import os 
import pandas as pd
import numpy as np

os.chdir('/home/t326h379/SUMO_2_3_Andrew')

Header_name = ["Amino_Acid"]

col_of_feature = [i for i in range(1,1025)]

Header_name = Header_name + col_of_feature

def open_file(PID, csv_file_name, Position_individual):  

    
    df = pd.read_csv(csv_file_name, header=None)
    
    df.columns = Header_name

    
        

    with open(csv_file_name) as outfile:
        feature_vector = []
        AA = []
        lines = outfile.readlines()
        for line in lines:
            list = line.split(",")
            AA.append(list[0])
           
        AA = ''.join(AA)
        Position_individual = int(Position_individual)
        
        
        index = Position_individual-1
        
        if (AA[index] == "K"):
        

            matrix = df.iloc[index]
            grabbed_matrix =  np.array(matrix)


            flatten_value = grabbed_matrix.ravel()

            feature_vector = flatten_value.tolist()


            listToStr = ','.join([str(elem) for elem in feature_vector])
            



            with open('SUMO_2_3_Positive_Prot_T5_feature.txt','a+') as outfile:
#                 outfile.write(label)
#                 outfile.write(",")
                
                outfile.write(PID)
                outfile.write(",")
                
                outfile.write(str(Position_individual))
                outfile.write(",")
                
#                 outfile.write(Sequences_individual)
#                 outfile.write(",")

                outfile.write(listToStr)
                outfile.write("\n")
        
        
with open("Ninety_Percent_Training_Protein_Positive_SUMO_2_3_site_After_30_Percent_CD_HIT.txt") as serena:
    for line in serena:
#         print(line)
        x = line.split()
        csv_file_name = x[0]+"_Prot_Trans_.csv"
        PID = x[0]
        Position_individual = x[1].strip("\n")
        print(PID, csv_file_name, Position_individual)
        try:
            open_file(PID, csv_file_name, Position_individual)
        except:
            pass
```
